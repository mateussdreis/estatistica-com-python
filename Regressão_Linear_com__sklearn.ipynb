```python
plt.figure()
plt.scatter(dados_imoveis['area'], dados_imoveis['valor'])
plt.xlabel('√Årea do Im√≥vel')
plt.ylabel('Valor por metro quadrado')
plt.title('Dispers√£o dos dados')
plt.show()
```

```python
# Bibliotecas necess√°rias para fazer a regress√£o
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
X = dados_imoveis['area'].values

y = dados_imoveis['valor'].values
```

```python
X.shape # verificando o formato
```

```python
X = X.reshape(-1,1)
X.shape
```

```python
y.shape # verificando o formato da vari√°vel
```

```python
y = y.reshape(-1,1) # refazendo de maneira a ficar compat√≠vel com o modelo
```

```python
from sklearn.model_selection import train_test_split #biblioteca para fazer a valida√ß√£o cruzada
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) #separando em vari√°veis de teste e de treino
```

```python
regressor = LinearRegression() # criando o modelo

regressor.fit(X_train, y_train) # treinando o modelo
```

```python
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
plt.scatter(X_train, y_train,color='r')

plt.plot(X_test, y_pred,color='k')



plt.show()
```

```python
from sklearn.linear_model import LinearRegression
# Creating a Linear Regression model on our data
lin = LinearRegression()
lin.fit(dados_imoveis[['area']], dados_imoveis['valor'])
# Creating a plot
ax = dados_imoveis.plot.scatter(x='area', y='valor', c='k')
ax.plot(dados_imoveis[['area']], lin.predict(dados_imoveis[['area']]), c='r')
```
